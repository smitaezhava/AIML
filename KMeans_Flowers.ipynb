```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pylab as plt
```

```python
iris=pd.read_csv('IRIS.csv')
```

```python
iris.info()
```

```python
iris_ind_attr=iris.drop('species', axis=1)
```

```python
iris_ind_attr.head()
```

```python
sns.pairplot(iris_ind_attr,diag_kind='kde')
```

```python
clusters=range(1,10)
meanDistortions=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(iris_ind_attr)
    prediction=model.predict(iris_ind_attr)
    #print(prediction)
    meanDistortions.append(model.inertia_)#(sum(np.min(cdist(tech_sup,model.cluster_centers_))))
```

```python
plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average Distortion')
plt.title('K with Elbow')
```
